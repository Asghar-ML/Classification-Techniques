# Importing Required Libraries

```python
#import modules
import pandas as pd # for dataframes
import datetime as dt
```

# Importing Dataset

```python
data = pd.read_csv("Taxi_Trips.csv")
```

```python
data.head()
```

# Data Pre-Processing

```python
data.info()
```

## Typecasting

```python
data['Trip End Timestp'] = pd.to_datetime(data['Trip End Timestp'])
```

```python
data.info()
```

# Filter Required Columns

```python
data=data[['Taxi ID','Trip End Timestp','Trip Total','Company']]
data.head()
```

```python
data['Trip End Timestp'].min(), data['Trip End Timestp'].max()
```

```python
PRESENT = dt.datetime(2019,9,28)
print(PRESENT)
```

# RFM Analysis

```python
rfm= data.groupby('Company').agg({'Trip End Timestp': lambda date: (PRESENT - date.max()).days,'Taxi ID': lambda num: len(num),'Trip Total': lambda price: price.sum()})
rfm.columns
```

### Rename Columns

```python
rfm.columns=['recency','frequency','monetary']
rfm['recency'] = rfm['recency'].astype(int)
rfm.head()
```

### Computing Quantile fo RFM Values

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
rfm.head()
```

### Scoring for Interpreting Result

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```

# Top 10 Company

```python
rfm[rfm['RFM_Score']=='111'].sort_values('monetary', ascending=False).head(10)
```
