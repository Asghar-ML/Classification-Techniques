## Filtering Required Columns

```python
df=data[['Taxi ID','Trip End Timestp','Trip Total','Trip ID']]
df.head()
```

### Checking Range of Timestamp

```python
df['Trip End Timestp'].min(), df['Trip End Timestp'].max()
```

```python
PRESENT = dt.datetime(2019,9,3)
df['Trip End Timestp'] = pd.to_datetime(df['Trip End Timestp'])
df.head()
```

# RFM Analysis

## Grouping Featrures

```python
rfm= df.groupby('Taxi ID').agg({'Trip End Timestp': lambda date: (PRESENT - date.max()).days,'Trip ID': lambda num: len(num),'Trip Total': lambda price: price.sum()})
rfm.columns
```

```python
rfm = rfm.fillna(0)
rfm.columns=['recency','frequency','monetary']
rfm['recency'] = rfm['recency'].astype(int)
rfm.head()
```

## Dividing in Quartiles

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
rfm.head()
```

## Generating RFM Score 

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```

## Result

```python
rfm[rfm['RFM_Score']=='111'].sort_values('monetary', ascending=False).head(10)
```
